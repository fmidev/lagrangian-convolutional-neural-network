```python
import argparse
import os
import yaml
import shutil

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from ipywidgets import widgets

exn = "extrapolation"
lcnn = "lcnn-diff-rmse-30lt-20062022"
ln = "linda"
rn = "t11-rn-logcosh-lt30"
```

# Continuous scores

```python
# Read scores from csv
csvfn = f"/mnt/data/PINCAST/manuscript_1/verification/lcnn/lcnn-logcosh-21082022-figs-article/CONT_values.csv"
df = pd.read_csv(csvfn, index_col=[0, 1])

df = df.T

df.head()
```

## MAE

```python
var = "MAE"
df_mae = df[var].loc[df.index.astype(int) <= 12]

diff_mae_extrap = (df_mae[lcnn] - df_mae[exn]).to_frame().copy() 
diff_mae_linda = (df_mae[lcnn] - df_mae[ln]).to_frame().copy() 
diff_mae_rainnet = (df_mae[lcnn] - df_mae[rn]).to_frame().copy() 

diff_mae_extrap["relative"] = (df_mae[lcnn] - df_mae[exn]) / np.abs(df_mae[exn]) * 100
diff_mae_linda["relative"] = (df_mae[lcnn] - df_mae[ln]) / np.abs(df_mae[ln]) * 100
diff_mae_rainnet["relative"] = (df_mae[lcnn] - df_mae[rn]) / np.abs(df_mae[rn]) * 100

print(f"{var} DIFFERENCE")
print(f"Difference in {var}: LCNN vs extrapolation")
print(diff_mae_extrap)
print(f"Max difference at {diff_mae_extrap[0].argmax()+1} {np.max(diff_mae_extrap[0])}")

print(f"Difference in {var}: LCNN vs LINDA")
print(diff_mae_linda)
print(f"Max difference at {diff_mae_linda[0].argmax()+1}: {np.max(diff_mae_linda[0])}")

print(f"Difference in {var}: LCNN vs RainNet")
print(diff_mae_rainnet)
print(f"Max difference at {np.argmax(diff_mae_rainnet[0]) +1}: {np.max(diff_mae_rainnet[0])}")
```

## ME

```python
var = "ME"
df_mae = df[var].loc[df.index.astype(int) <= 12]

diff_mae_extrap = (df_mae[lcnn].abs() - df_mae[exn].abs()).to_frame().copy() 
diff_mae_linda = (df_mae[lcnn].abs() - df_mae[ln].abs()).to_frame().copy() 
diff_mae_rainnet = (df_mae[lcnn].abs() - df_mae[rn].abs()).to_frame().copy() 

diff_mae_extrap["relative"] = (df_mae[lcnn].abs() - df_mae[exn].abs()) / np.abs(df_mae[exn]) * 100
diff_mae_linda["relative"] = (df_mae[lcnn].abs() - df_mae[ln].abs()) / np.abs(df_mae[ln]) * 100
diff_mae_rainnet["relative"] = (df_mae[lcnn].abs() - df_mae[rn].abs()) / np.abs(df_mae[rn]) * 100

print(f"{var} DIFFERENCE")
print(f"Difference in {var}: LCNN vs extrapolation")
print(diff_mae_extrap)
print(f"Max difference at {diff_mae_extrap[0].argmax()+1} {np.max(diff_mae_extrap[0])}")

print(f"Difference in {var}: LCNN vs LINDA")
print(diff_mae_linda)
print(f"Max difference at {diff_mae_linda[0].argmax()+1}: {np.max(diff_mae_linda[0])}")

print(f"Difference in {var}: LCNN vs RainNet")
print(diff_mae_rainnet)
print(f"Max difference at {np.argmax(diff_mae_rainnet[0]) +1}: {np.max(diff_mae_rainnet[0])}")
```
